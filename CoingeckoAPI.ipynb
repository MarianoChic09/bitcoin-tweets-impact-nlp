```python
top_20=coin_market.id[:21]
top_20
```

```python
fecha_inicio=dt.datetime(day=11,month=3,year=2022).timestamp()
fecha_fin=dt.datetime(day=14,month=3,year=2022).timestamp()

df=pd.DataFrame(cg.get_coin_market_chart_range_by_id(id='bitcoin',vs_currency='usd',from_timestamp=fecha_inicio,to_timestamp=fecha_fin))

df['id']=top_20[0]
for coin in top_20[1:21]:
    df2=df
    df_aux=pd.DataFrame(cg.get_coin_market_chart_range_by_id(id=coin,vs_currency='usd',from_timestamp=fecha_inicio,to_timestamp=fecha_fin))
    df_aux['id']=coin
    df=pd.concat([df2,df_aux],ignore_index=True,axis=0)
    #print(df.head(2))

    
```

```python
df2=df.drop('id',axis=1).copy()

for column in df.drop(['id'],axis=1).columns:
    df2[column]=df[column].apply(lambda x:x[1])
    df2['Timestamp']=df[column].apply(lambda x:x[0])

df2['id']=df['id']
df2['Timestamp2']=df2['Timestamp'].apply(lambda x: dt.datetime.fromtimestamp(x/1000).strftime('%Y-%m-%d %H:%M:%S'))
```
